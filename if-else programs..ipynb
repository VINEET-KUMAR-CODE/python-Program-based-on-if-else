# 20 practical Python questions based on if-else statements.

```python
import pandas as pd
import numpy as np
```

## 1. Discount Calculator

```python
price = float(input("Enter product price: "))

if price > 500:
    discount = price * 0.10
else:
    discount = price * 0.05

final_price = price - discount
print("Final Price:", final_price)
```

## 2. Driving Eligibility

```python
age = int(input("Enter your age: "))

if age >= 18:
    print("Eligible to drive")
else:
    print("Not eligible to drive")
```

## 3. Odd or Even Age Check

```python
age = int(input("Enter your age: "))

if age % 2 == 0:
    print("Even age")
else:
    print("Odd age")
```

## 4. Store Timing

```python
time = int(input("Enter time in 24-hour format: "))

if 9 <= time <= 21:
    print("Store is open")
else:
    print("Store is closed")
```

## 5. Pass or Fail

```python
marks = int(input("Enter your marks: "))

if marks >= 40:
    print("Pass")
else:
    print("Fail")
```

## 6. Prime Membership

```python
prime = input("Do you have Prime membership? (yes/no): ")

if prime.lower() == "yes":
    print("Free delivery")
else:
    print("Delivery charge: ₹50")
```

## 7. Leap Year Checker

```python
year = int(input("Enter year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap year")
else:
    print("Not a leap year")
```

## 8. Temperature Alert

```python
temp = float(input("Enter temperature in °C: "))

if temp > 40:
    print("Heat Alert")
elif temp < 0:
    print("Cold Alert")
else:
    print("Normal Weather")
```

## 9. Grade Calculator

```python
percent = float(input("Enter percentage: "))

if percent >= 90:
    print("Grade: A+")
elif percent >= 80:
    print("Grade: A")
elif percent >= 70:
    print("Grade: B")
else:
    print("Grade: C")
```

## 10. Bank Withdrawal

```python
balance = float(input("Enter account balance: "))
withdraw = float(input("Enter withdrawal amount: "))

if withdraw <= balance:
    print("Withdrawal successful")
else:
    print("Insufficient balance")
```

## 11. Number Comparison

```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

if a > b:
    print("First number is greater")
elif a < b:
    print("First number is smaller")
else:
    print("Both numbers are equal")
```

## 12. Password Validator

```python
password = input("Enter password: ")

if password == "secure123":
    print("Access granted")
else:
    print("Access denied")
```

## 13. Evening Discount

```python
time = int(input("Enter current time (24-hour): "))
price = float(input("Enter product price: "))

if 18 <= time <= 21:
    discount = price * 0.20
else:
    discount = 0

final_price = price - discount
print("Final Price:", final_price)
```

## 14. Eligible for Voting

```python
citizenship = input("Enter citizenship: ")
age = int(input("Enter age: "))

if citizenship.lower() == "indian" and age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible to vote")
```

## 15. Restaurant Bill Split

```python
people = int(input("Enter number of people: "))
bill = float(input("Enter total bill: "))

if people > 5:
    service_charge = bill * 0.15
else:
    service_charge = 0

final_bill = bill + service_charge
print("Final Bill Amount:", final_bill)
```

## 16. Delivery Service

```python
location = input("Enter delivery location (urban/rural): ")

if location.lower() == "urban":
    print("Delivery available")
else:
    print("Delivery not available")
```

## 17. BMI Calculator

```python
weight = float(input("Enter weight (kg): "))
height = float(input("Enter height (m): "))

bmi = weight / (height * height)
print("BMI:", bmi)

if bmi < 18.5:
    print("Underweight")
elif bmi <= 24.9:
    print("Normal")
else:
    print("Overweight")
```

## 18. Weekend Check

```python
day = input("Enter day of the week: ")

if day.lower() in ["saturday", "sunday"]:
    print("Weekend")
else:
    print("Weekday")
```

## 19. Flight Fare Check

```python
age = int(input("Enter age: "))
fare = float(input("Enter ticket fare: "))

if age < 12 or age > 60:
    final_fare = fare * 0.50
else:
    final_fare = fare

print("Final Fare:", final_fare)
```

# 20. Electricity Bill

```python
units = int(input("Enter electricity units: "))

if units <= 100:
    bill = units * 5
elif units <= 300:
    bill = units * 10
else:
    bill = units * 15

print("Total Bill:", bill)
```
